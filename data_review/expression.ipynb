```python
import pandas as pd

# 读取整个CSV文件
df = pd.read_csv('../data/GeneExpression/OmicsExpressionProteinCodingGenesTPMLogp1.csv')

# 查看前5行
print(df.head())
```

```python
dict = {"K562": 'ACH-000551', 
        "JURKAT": 'ACH-000995', 
        "A549": "ACH-000681",
        "MEL202": "ACH-001554",
        "PK1": "ACH-000307",
        "PATU8988S":"ACH-000022",
        "IPC298": "ACH-000915",
        "HSC5":"ACH-001524",
        "HS936T": "ACH-000801",
        "MELJUSO":"ACH-000881"}
```

```python
import pandas as pd
import pickle

for key in dict:
    c_data = df[df['Unnamed: 0'] == dict[key]]
    if not c_data.empty:
        c_row = c_data.drop(columns=['Unnamed: 0'])

        gene_expression_dict = {
            col.split(' ')[0]: c_row.iloc[0][col]
            for col in c_row.columns
        }

        with open(f'../data/GeneExpression/expr_{key}.pkl', 'wb') as f:
            pickle.dump(gene_expression_dict, f)

        print("保存完成，包含", len(gene_expression_dict), "个清洗后的基因表达值。")
    else:
        print(f"not found {key}")
```

```python
import pandas as pd
import pickle

for key in dict:
    # 读取 gene_expression.pkl（或你生成的 cleaned 版本）
    with open(f'../data/GeneExpression/expr_{key}.pkl', 'rb') as f:
        gene_expr_dict = pickle.load(f)

    # 读取 SL 数据（例如从 CSV 文件）
    sl_df = pd.read_csv(f'../data/SL_data/SLKB_cellline/SLKB_{key}.csv')  # 替换为实际文件路径

    # 统计总数和覆盖数
    missing_gene_pairs = []
    missing_genes = set()
    total_gene = 0
    total_pair = 0

    # 遍历 SL 数据的每一对基因
    for _, row in sl_df.iterrows():
        total_pair += 1
        g1 = row['gene_1']
        g2 = row['gene_2']
        g1_missing = g1 not in gene_expr_dict
        g2_missing = g2 not in gene_expr_dict
        if g1_missing or g2_missing:
            missing_gene_pairs.append((g1, g2))
            if g1_missing:
                missing_genes.add(g1)
            if g2_missing:
                missing_genes.add(g2)

    # 输出结果
    print(key)
    print(f"缺失的基因对数量: {len(missing_gene_pairs)}")
    print(f"缺失的基因数量: {len(missing_genes)}")
    print(len(missing_gene_pairs)/total_pair)
    print("\n部分缺失基因（前10个）:", list(missing_genes)[:10])
```
